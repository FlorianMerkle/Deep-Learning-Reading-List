```python
import torch, torchvision
import torch.nn as nn
import torch.nn.functional as F
import torchvision.transforms.functional as TF
import random
import matplotlib.pyplot as plt
device = "cuda"
```

# GAN FROM SCRATCH:

1st network: Generative: 
- Input =  Seed, e.g. torch.rand(100)
- Output = Tensor with shape (224,224,3)
- Loss Function = Negative loss of discriminator model

2nd Network: Discriminative
- Input = Image or output tensor of 1st network with shape (224,224,3)
- Output = Indicator if natural or generated input with shape (1)
- Loss Function = Binary Cross Entropy with input(0/1 for natural/generated, and discriminator output for natural images and generated images)

# Helper Modules

```python
class StdConv(torch.nn.Module):
    def __init__(self, in_c, out_c, activation='relu', stride=2, padding=0):
        super().__init__()
        self.conv = nn.Conv2d(in_c, out_c,kernel_size=3, stride=stride, padding=padding)
        self.bn = nn.BatchNorm2d(out_c)
        self.a = activation
        
    def forward(self, x):
        x = self.conv(x)
        if self.a=='leaky relu': x = F.leaky_relu(x, .2)
        if self.a=='relu': x = F.relu(x)
        x = self.bn(x)
        return x

class Upsample(torch.nn.Module):
    def __init__(self, in_c, out_c):
        super().__init__()
        self.convT = nn.ConvTranspose2d(in_c, out_c,
                                       kernel_size=3, stride=2,
                                       padding=0, output_padding=0)
        self.bn = nn.BatchNorm2d(out_c)

    def forward(self, x):
        x = self.convT(x)
        x = F.relu(x)
        x = self.bn(x)
        return x
    
```

# Generator Model

```python
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.lin = nn.Linear(100, 784)
        
        self.upsample1 = Upsample(1, 32)
        self.upsample2 = Upsample(64, 128)
        self.upsample3 = Upsample(128, 64)
        
        self.conv1 = StdConv(32, 64)
        self.conv2 = StdConv(128, 128)
        self.conv3 = StdConv(64, 1)
    
    def forward(self,inputs):
        x = self.lin(inputs)
        x = x.view(x.shape[0],1,28,28)
        x = self.upsample1(x)
        x = self.conv1(x)
        x = self.upsample2(x)
        x = self.conv2(x)
        x = self.upsample3(x)
        x = self.conv3(x)
        return torch.tanh(TF.center_crop(x,(28,28)))
```

# Discriminator

```python
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = StdConv(1,64, activation='leaky relu')
        self.conv2 = StdConv(64,128, activation='leaky relu')
        self.conv3 = StdConv(128,32, activation='leaky relu')
        self.lin = nn.Linear(128, 1)

    def forward(self, inputs):
        x = self.conv1(inputs)
        x = self.conv2(x)
        x = self.conv3(x)
        x = x.view(x.shape[0],x.shape[1]*x.shape[2]*x.shape[3])
        x = self.lin(x)
        return x
```

# Data

```python
ROOT = '/home/flo/data/mnist'
transform=torchvision.transforms.Compose([
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize((0.5,), (0.5))
])

train_ds = torchvision.datasets.MNIST(ROOT, download=True, transform=transform)
BS = 32
train_dl = torch.utils.data.DataLoader(train_ds, batch_size=BS, drop_last=True)

def denormalize(img):
    if len(img.shape)==2: img = img[None, None, :,:]
    img = TF.normalize(img, (-0.5/0.5), (1/.5))
    return img
```

# Training

### Training Procedure
for batch in data loader:
batch shape = e.g. 56,1,28,28  
generate data with generator  
output generator e.g. 56,1,28,28  
append labels for discriminator --> 0 for natural, 1 for generated  
discriminator predicts mixed data  
calculate BCE loss for discriminator  

### update rule for the discriminator:  
update the weights with the gradients of the weights calculated w.r.t. to the BCE  

### goal of the generator:  
maximize the loss of the discriminator  

### update rule for generator  
update the weights with the gradients of the loss function wrt to the discriminators loss (maximise loss)   



### Simple Logger Class

```python
import pandas as pd
class gan_logger():
    def __init__(self):
        self.stats = pd.DataFrame(columns=['epoch','d_loss','g_loss'])
        self.images = {}
    def log(self, epoch, d_loss, g_loss):
        row = {'epoch': epoch, 'd_loss': d_loss, 'g_loss': g_loss}
        self.stats = self.stats.append(row, ignore_index=True)
    def create_images(self, e, g, device):
        test_inp = torch.randn(10,100).to(device)
        imgs = g(test_inp)
        self.images[e] = imgs
        
```

### Training Procedure

```python
EPOCHS = 250        
logger = gan_logger()
loss = torch.nn.BCEWithLogitsLoss()
d = Discriminator().to(device)
g = Generator().to(device)
lr = 3e-4
d_optim = torch.optim.Adam(d.parameters(), lr=lr)
g_optim = torch.optim.Adam(g.parameters(), lr=lr)
for e in range(EPOCHS):
    print('Epoch:', e)
    acc_d_loss = 0
    acc_g_loss = 0
    for i, batch in enumerate(train_dl):
        
        #create noisy labels
        valid_labels = (torch.zeros(BS,1)+torch.Tensor(BS,1).uniform_(0,.25)).to(device)
        fake_labels = (torch.ones(BS,1)+torch.Tensor(BS,1).uniform_(-.25,0)).to(device)
        
        #Train Generator
        inp = torch.randn(BS,100).to(device)
        fake_imgs = g(inp)
        preds = d(fake_imgs)

        #measure the ability of the generator to pretend the images are real
        g_loss = loss(d(fake_imgs), valid_labels)
        g_optim.zero_grad()
        g_loss.backward()
        g_optim.step()
        acc_g_loss +=g_loss.item()
        
        #Train Discriminator
        real_imgs, _ = batch
        real_imgs = real_imgs.to(device)
        d_optim.zero_grad()
        inp = torch.rand(BS,100).to(device)
        evaluation_of_fake_images = d(g(inp)+.05*torch.randn_like(real_imgs))

        #measure the ability of the discriminator to classify real images as real and fake as fake
        img_noise = torch.randn_like(real_imgs)
        d_loss = (loss(d(real_imgs+.05*torch.randn_like(real_imgs)), valid_labels) + loss(evaluation_of_fake_images, fake_labels))/2
        acc_d_loss +=d_loss.item()
        d_loss.backward()
        d_optim.step()
        if i==0:
            with torch.no_grad():
                ratio_of_detected_fakes = 1 - evaluation_of_fake_images.sigmoid().round().sum().item()/len(inp)
                if ratio_of_detected_fakes < .3: skip_generator_training = True
                if ratio_of_detected_fakes > .3: skip_generator_training = False
                print('ratio of detected fake images: ', ratio_of_detected_fakes)
                
    epoch_d_loss = acc_d_loss/len(train_dl)
    epoch_g_loss = acc_g_loss/len(train_dl)
    logger.log(e, epoch_d_loss, epoch_g_loss)
    logger.create_images(e, g, device)
    print(len(logger.images.values()))
    print('discriminator loss: ',epoch_d_loss)
    print('generator loss: ', epoch_d_loss)
```
